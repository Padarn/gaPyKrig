```python
import math
import operator
import decimal
import scipy.spatial as spatial

#From the internet
def area(p):
    return 0.5 * abs(np.sum(x0*y1 - x1*y0
                         for ((x0, y0), (x1, y1)) in segments(p)))

def segments(p):
    return zip(p, p[1:] + [p[0]])

def in_hull(p, hull):
    if not isinstance(hull,spatial.Delaunay):
        hull = spatial.Delaunay(hull)
    
        return hull.find_simplex(p)>=0

#From Padarn
def convex_hull(points):
    """Computes the convex hull of a set of 2D points.

    Input: an iterable sequence of (x, y) pairs representing the points.
    Output: a list of vertices of the convex hull in counter-clockwise order,
      starting from the vertex with the lexicographically smallest coordinates.
    Implements Andrew's monotone chain algorithm. O(n log n) complexity.
    """

    # Sort the points lexicographically (tuples are compared lexicographically).
    # Remove duplicates to detect the case we have just one unique point.
    points = sorted(set(points))

    # Boring case: no points or a single point, possibly repeated multiple times.
    if len(points) <= 1:
        return points

    # 2D cross product of OA and OB vectors, i.e. z-component of their 3D cross product.
    # Returns a positive value, if OAB makes a counter-clockwise turn,
    # negative for clockwise turn, and zero if the points are collinear.
    def cross(o, a, b):
        return (a[0] - o[0]) * (b[1] - o[1]) - (a[1] - o[1]) * (b[0] - o[0])

    # Build lower hull 
    lower = []
    for p in points:
        while len(lower) >= 2 and cross(lower[-2], lower[-1], p) <= 0:
            lower.pop()
        lower.append(p)

    # Build upper hull
    upper = []
    for p in reversed(points):
        while len(upper) >= 2 and cross(upper[-2], upper[-1], p) <= 0:
            upper.pop()
        upper.append(p)

    # Concatenation of the lower and upper hulls gives the convex hull.
    # Last point of each list is omitted because it is repeated at the beginning of the other list. 
    return lower[:-1] + upper[:-1]


# Example: convex hull of a 10-by-10 grid.
#assert convex_hull([(i//10, i%10) for i in range(100)]) == [(0, 0), (9, 0), (9, 9), (0, 9)]
```

```python
for survey in list(df):
    df[survey]['natlog_flux_g/m2/d'] =  df[survey]['converted_flux_g/m2/d'].apply(np.log)
    #Lewicki MVUE 
#    ybar_MVU = df[survey]['log_flux_log(g/m2/d)'].mean() #arithemtic mean of log transformed values
#    var_MVU  = df[survey]['log_flux_log(g/m2/d)'].var() #Variance of transformed values
    ybar_MVU = df[survey]['natlog_flux_g/m2/d'].mean() #arithemtic mean of log transformed values
    var_MVU  = df[survey]['natlog_flux_g/m2/d'].var() #Variance of transformed values
    n = len(df[survey]['log_flux_log(g/m2/d)'])
    mu_MVU = math.exp(ybar_MVU+(var_MVU/2)) #Mean of lognormal distribution
    sigma2_MVU = (mu_MVU**2) * (math.exp(var_MVU)-1) #Varaiacne of log normal distribution
    
    #Elio MVUE method
    sichel=0
    j=2
    A=1
    while A>0.000001:
        A =  ((n-1)**(2*j-1))*var_MVU**j /((n**j * reduce(operator.mul,range(n+2*j-1,n-1,-2)))*math.factorial(j)) # Padarn found mistak, rather than n^j + ... it should be n^j *xxx
        sichel += A
        j+=1
    mu_MVU_Elio = (1 +((n-1)/n)*var_MVU + sichel) * math.exp(ybar_MVU)

    #Lewicki AM
    ybar_AM = df[survey]['converted_flux_g/m2/d'].mean() #arithemtic mean of log transformed values
    var_AM  = df[survey]['converted_flux_g/m2/d'].var()
    

    #Emission rate (gross)
    xy_data = np.column_stack((df[survey]['x_m'],df[survey]['y_m']))
    xy_draft = tuple(map(tuple,xy_data))
    convexpts = convex_hull(xy_draft)
    #convex_hull = spatial.qhull.Delaunay(xy_data).convex_hull
    surfarea = area(convexpts)
    
    #Convert to kg/d
    AM_rate = ybar_AM*surfarea/1000
    MVU_Elrate =  mu_MVU_Elio*surfarea/1000
    MVU_Lewrate = mu_MVU*surfarea/1000
    
    Resdf.ix[survey, 'AM_rate'] = AM_rate
    Resdf.ix[survey, 'MVU_Elrate'] = MVU_Elrate
    Resdf.ix[survey, 'MVU_Lewrate'] = MVU_Lewrate
    Resdf.ix[survey, 'surface_area'] = surfarea
    
    ######################## PLOTTING ###########################
    plotting_on = True
    plot_hull = True
    #############################################################
    
    """
    File Input - Define filenames and variable names.
    """
    
    ## Data Options
    filename = 'Ginninderra 2013 soil flux surveys-blanks deleted.xls' # Name of your .xls file
    sheetname = 'Jan24 for GIS' # Name of the sheet to use
    
    xdata = df[survey].x_m.values
    ydata = df[survey].y_m.values
    flux = df[survey]['converted_flux_g/m2/d'] /1000 #To turn into kg/m2/d
        
    """ 
    Domain - Set up the domain of the intergral. 
    
    This defines both plotting and
    the maximum extent of the grid for the flux integral. The boundary of the 
    domain is estimated by using the mean of the background.
    
    requires: domain = [xmin,xmax,ymin,ymax]
    
    setting any of these quantities to 'None' will default to using the maximum
    and minimum coordinates read from the data
    
    Maximum possible domain is the convex hull.
    """
    domain = [None,None,None,None]
    #domain = [0, 120, -30, 0]
    
    """
    Background - Determine which points belong to 'background'
            constant_background = True/False
        determined if background is constant or linear 
    """
    
    constant_background = True
    
    """
    Points can be specified to be background in two ways, 'outside box exclusion'
    and 'neighbourhood' exclusion.
    """
    
    
    """
    1) outside box exlusion - specify a box by defining the list 
                    exlusion_box = [xmin,xmax,ymin,ymax]
       then all points lying outside of this box are included in the background
       NOTE - points on boundary of box are not part of the background
    """
    
    exclusion_box = [None,None,None,None]
    #exclusion_box = [0,120,-9.9,9.9]
    """
    2) neighbourhood exclusion - specify points and small neighbourhood to include
       or exclude from the above boxed background points. 
       specify a two lits with elements with form:
                        (xcenter,ycenter,distance)
       e.g
                               (1, 1, 0.5)
       Adding this element to the 'include_neighbourhood' specifies that points 
       within 0.5m of the location x=1m, y=1m, will be included in the background.
    """
    
    include_neighbourhood = []  
    exclude_neighbourhood = [] #Oct2, Jan24
    #exclude_neighbourhood = [(36.8,-10,0.5),(103.5,10,0.5)] #Oct  15
    #exclude_neighbourhood = [(34.5,10,0.5),(39.1,10,0.5),(34.5,-10,0.5)] #Oct24
    #exclude_neighbourhood = [(39.1,10,0.5)] #Oct31
    #exclude_neighbourhood = [(39.1,10,0.5),(39.1,-10,0.5)] #Nov6, Dec 20
    #exclude_neighbourhood = [(59.8,-10,0.5),(48.3,-10,0.5),(48.3,10,0.5),(39.1,-10,0.5),(39.1,10,0.5),(29.9,-20,0.5)] #Nov 14
    #exclude_neighbourhood = [(39.1,10,0.5),(39.1,-10,0.5),(48.3,-10,0.5)] #Nov27
    #exclude_neighbourhood = [(39.1,10,0.5),(39.1,-10,0.5),(59.8,10,0.5),(59.8,30,0.5)] #Dec4
    #exclude_neighbourhood = [(39.1,10,0.5),(39.1,-10,0.5),(115,10,0.5)] #Dec13
    
    """
    Extra options - Should leave alone without good reason
    """
    gridres = 1000 # integration grid resolution (should be fine)
    
    """
    ========================END SETUP=====================================
    """
    
    #--------------------- hull function --------------------------
    
    def in_hull(p, hull):
        if not isinstance(hull,spatial.Delaunay):
            hull = spatial.Delaunay(hull)
    
        return hull.find_simplex(p)>=0
    
    
    # ------------------- domain set up ----------------------------
    
    # for any domain coordinates not specified get from data
    if domain[0] is None:
        domain[0] = np.min(xdata)
    if domain[1] is None:
        domain[1] = np.max(xdata)
    if domain[2] is None:
        domain[2] = np.min(ydata)
    if domain[3] is None:
        domain[3] = np.max(ydata)
    xmin,xmax,ymin,ymax = domain
    xran = xmax-xmin
    yran = ymax-ymin
    
    # generate grids
    grid_x, grid_y = np.mgrid[xmin:xmax:xran*1.0/gridres,
                              ymin:ymax:yran*1.0/gridres]
    
    # ------------------ background setup ---------------------
    
    # exclusion box
    exclusion_box_rule = np.logical_or(xdata<exclusion_box[0],
                                        xdata>exclusion_box[1])
    exclusion_box_rule = np.logical_or(exclusion_box_rule,
                                        ydata<exclusion_box[2])
    exclusion_box_rule = np.logical_or(exclusion_box_rule,
                                        ydata>exclusion_box[3])
    
    # inclusion neighbourhoods
    neighbourhood_rules_include = np.zeros(len(xdata), dtype=bool)
    for (xc,yc,dist) in include_neighbourhood:
        rule = (xdata-xc)**2+(ydata-yc)**2 < dist
        print rule
        neighbourhood_rules_include = np.logical_or(neighbourhood_rules_include,rule)
    
    # exclusion neighbourhoods
    neighbourhood_rules_exclude = np.ones(len(xdata), dtype=bool)
    for (xc,yc,dist) in exclude_neighbourhood:
        rule = (xdata-xc)**2+(ydata-yc)**2 > dist
        neighbourhood_rules_exclude = np.logical_and(neighbourhood_rules_exclude,rule)
    
    background_rule = np.logical_or(exclusion_box_rule,neighbourhood_rules_include)
    background_rule = np.logical_and(background_rule,neighbourhood_rules_exclude)
    
    # background fux
    #bg_flux_average = np.mean(flux[background_rule])
    #Ivan changed for calulations to not have a background
    bg_flux_average = 0
    
    # --------------- add data to corners of domain -------------
    xy = np.column_stack((xdata,ydata))
    
    pts = set([(x[0],x[1]) for x in xy])
    if len(pts)!=len(xdata):
        print "WARNING: You have duplicate points in your data."
        print "Script will continue, but you may get strange results."
    
    ######################## INTEGRALS ##############################
    
    # ----------------  get pts and hull ----------------------
    
    xy_data = np.column_stack((xdata,ydata))
    xy_draft = tuple(map(tuple,xy_data))
    convexHull = convex_hull(xy_draft)
    
    
    xx = grid_x.reshape(gridres*gridres,)
    yy = grid_y.reshape(gridres*gridres,)
    gridpts = np.column_stack((xx,yy))
    inhull_test = in_hull(gridpts,xy_data)
    inhull_test = inhull_test.reshape(gridres,gridres)
    
    # ----------------  integration grid ---------------------
    
    grid_x_c = (grid_x[:-1,:-1]+grid_x[1:,:-1])/2
    grid_y_c = (grid_y[:-1,1:]+grid_y[1:,1:])/2
    
    dx = -grid_x[0,0]+grid_x[1,0]
    dy = -grid_y[0,0]+grid_y[0,1]
    dxdy = dx*dy
    
    # ----------------  calculate background integral ----------------
    if constant_background:
        def bg_func(grid_x,grid_y):
            return np.ones(grid_x.shape)*bg_flux_average
    else: 
        def bg_func(grid_x,grid_y):
            fluxinter =  interpolate.griddata(xy_data[background_rule],
                         flux[background_rule], (grid_x,grid_y), method='linear', fill_value = bg_flux_average)
            return fluxinter
    
    background_grid = bg_func(grid_x, grid_y)
    background_grid[np.logical_not(inhull_test)] = 0 #In_hull test replaces convex_hull
    
    background_c=(background_grid[:-1,:-1]+background_grid[1:,:-1]+
                 background_grid[:-1,1:]+background_grid[1:,1:])/4
    # If convexhull, set all points ouside convex hull of data to zero
    
    
    background_integral = np.sum(np.sum(dxdy*background_c))
    
    
    # ----------------  calculate flux integral ---------------
    
    def flux_func(grid_x, grid_y):
        fluxinter =  interpolate.griddata(xy_data, flux, 
                (grid_x,grid_y), method='linear', fill_value = bg_flux_average)
        bginter = bg_func(grid_x,grid_y)
        fluxinter[(fluxinter-bginter)<0] = bginter[(fluxinter-bginter)<0] #zero correctiom
        return fluxinter
    
    flux_grid = flux_func(grid_x,grid_y)
    flux_grid[np.logical_not(inhull_test)] = 0
    
    flux_c=(flux_grid[:-1,:-1]+flux_grid[1:,:-1]+
            flux_grid[:-1,1:]+flux_grid[1:,1:])/4
    
    
    flux_integral = np.sum(np.sum(dxdy*flux_c))
    
    
    #------------------ Area Integral ---------------------
    
    flux_c[flux_c!=0] = 1
    area_integral  = np.sum(np.sum(dxdy*flux_c))
    
    # ----------------- print results -------------------- 
    
    Resdf.ix[survey, 'Cubic_rate'] = flux_integral
```
